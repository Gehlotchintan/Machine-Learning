# How to create other plot with the help of iplot

```python
df.iplot('scatter',x='A',y='B',mode='markers')  #we wrte markers cuz if we dont then it will
                                                 #connect all the points in a line
```

```python
df2
```

```python
df2.iplot(kind='bar',x='Catagory',y='Values')
```

```python
df.iplot('bar')           #will get a bumch of bars thats why we nees to chpose some specifi coloumn
```

```python
#to count the intenses of each column
df.count().iplot('bar')
```

```python
#sum of each column
df.sum().iplot('bar')
```

```python
df.iplot('box')
```

```python
df3=pd.DataFrame({'x':[1,2,3,4,5],'y':[10,20,30,40,50],'z':[5,4,3,2,1]})
```

```python
df3
```

```python
#for 3D plots we need three variables as resides in df3
df3.iplot('surface')
```

```python
#we can change the color for different scale regon:
df3.iplot('surface',colorscale='rdylbu')
#keep in mind for blue we have 'bu'   keyword is 'colorscale'
```

```python
df['A'].iplot('hist',bins=34)
```

```python
#if yu dirrectly apply all the columns intead of only one then yu will get overlapping hists
df.iplot('hist')
```

```python
#spread type:usually use for stock data
df[['A','B']].iplot('spread')
```

```python
#bubble plot-similer to scatter but size of the bubble will be diffrent
df.iplot('bubble',x='A',y='B',size='C')
```

```python
#scatter matrix---its basically similer to the pairplot of seaborn library
df.scatter_matrix()
#itll take a lot of time for large datasets and also can crash the libraray sobe aware
```
