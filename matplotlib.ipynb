```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x=np.linspace(0,5,11)
y=x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y,'r')
plt.xlabel('Force')
plt.ylabel('Distence')
plt.title('Work done')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('x label')
axes.set_ylabel('y label')
axes.set_title('title')
```

```python
axes.plot(x,y)
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y)
axes2.plot(y,x)
axes1.set_title('arm')
axes2.set_title('ram')
```

```python
fig,axes=plt.subplots()
fig,axes=plt.subplots(nrows=3,ncols=3)
plt.tight_layout()
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
for curr_ax in axes:
    curr_ax.plot(x,y)
    axes[0].plot(x,y)
    axes[1].plot(y,x)
   
```

```python
fig=plt.figure(figsize=(9,2),dpi=100)
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes=plt.subplots(nrows=2,ncols=1,figsize=(8,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.point(y,x)
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='orange',lw=3,ls='-',marker='o',markersize=20,markeredgewidth=3,markeredgecolor='green')
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple',lw=3,ls='--')
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
